# Data Cleaning using Scikit learn

## Method 1: we are going to use SimpleImputer class of Sklearn

```python
import numpy as np
import pandas as pd 
from sklearn.impute import SimpleImputer
```

```python
#read Data using pandas
train=pd.read_csv("./dataset/train.csv")
train
```

```python
#read Data using pandas
test=pd.read_csv("./dataset/test.csv")
test
```

```python
print("train shape=",train.shape)
print("testn shape=",test.shape)
```

```python
test.head()
```

```python
train.head()
```

```python
X_train=train.drop(columns="SalePrice")
y_train=train["SalePrice"]
print("shape of X_train df = ",X_train.shape)
print("shape of y_train df = ",y_train.shape)
```

### Numerical Missing Value Imputation

```python
#select columns which has numrical values
num_vars=X_train.select_dtypes(include=["int64","float64"]).columns
num_vars
```

```python
#print column wise null values
X_train[num_vars].isnull().sum()
```

```python
#create object of simple imputer 

imputer_mean = SimpleImputer(strategy='mean')
```

#### use ( stratergy=mean,mode,median ) when you have numrical data
#### use ( stratergy=constant ) when you have categorical data or when you want enter constant value manually

```python
#fit it to train data
imputer_mean.fit(X_train[num_vars])
```

```python
#after fitting get mean stats of each numerical colums
imputer_mean.statistics_
```

```python
#now apply this method to all columns which has numrical NAN values
imputer_mean.transform(X_train[num_vars])
```

```python
# we have got array in output but we want dataframe so
X_train[num_vars] = imputer_mean.transform(X_train[num_vars])
```

```python
#apply same for test data
test[num_vars] = imputer_mean.transform(test[num_vars])
```

```python
X_train[num_vars].isnull().sum()
```

```python
test[num_vars].isnull().sum()
```

### Categorical Missing Value Imputation

```python
#select columns which has categorical values
cat_vars=X_train.select_dtypes(include=["object"]).columns
cat_vars
```

```python
X_train[cat_vars].isnull().sum()
```

```python
#again creae an object and change statergy to most frequent word
imputer_mode = SimpleImputer(strategy='most_frequent')
```

```python
imputer_mode
```

```python
#fit it to train data
imputer_mode.fit(X_train[cat_vars])
```

```python
#now apply this method to all columns which has categorical NAN values
imputer_mode.statistics_
```

```python
#change transform method to imputer_mode
X_train[cat_vars] = imputer_mode.transform(X_train[cat_vars])
test[cat_vars] = imputer_mode.transform(test[cat_vars])
```

```python
X_train[cat_vars].isnull().sum()
```

```python
#check Null value in final train dataset
X_train.isnull().sum().sum()
```

```python
#check Null value in final test dataset
test.isnull().sum().sum()
```

```python
#all data is cleaned now
```

## Method 2 : Value imputation by creating Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
#read Data using pandas
train2=pd.read_csv("./dataset/train.csv")
train2
```

```python
#read Data using pandas
test2=pd.read_csv("./dataset/test.csv")
test2
```

```python
X_train2=train2.drop(columns="SalePrice")
y_train2=train2["SalePrice"]
print("shape of X_train df = ",X_train.shape)
print("shape of y_train df = ",y_train.shape)
```

```python
#select columns which has numrical values
num_vars2=X_train.select_dtypes(include=["int64","float64"]).columns
num_vars2
```

```python
#print column wise null values
X_train2[num_vars2].isnull().sum()
```

```python
#select columns which has categorical values
cat_vars2=X_train.select_dtypes(include=["object"]).columns
cat_vars2
```

```python
X_train2[cat_vars2].isnull().sum()
```

```python
#percentage of Nan Categorical values
isnull_per=X_train2[cat_vars2].isnull().mean()*100
isnull_per
```

```python
#select columns which have NaN greater than 0%
miss_vars=isnull_per[isnull_per>0].keys()
miss_vars
```

```python
#Randamly Select Columns 
```

```python
#Fill values with mean for particular column
num_var_mean=['LotFrontage']
#Fill values with median for particular columns
num_var_median=['MasVnrArea','GarageYrBlt']
#Fill values with mode for particular columns
cat_var_mode=['Alley', 'MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure','BsmtFinType1', 'BsmtFinType2', 'Electrical', 'FireplaceQu',]
#Fill values with some constant for particular columns
cat_var_const=['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PoolQC','Fence', 'MiscFeature']
```

#### Create Pipeline

```python
#assign data cleaning statergy and create pipeline
```

```python
num_var_mean_imputer=Pipeline(steps=[("imputer",SimpleImputer(strategy="mean"))])
num_var_median_imputer=Pipeline(steps=[("imputer",SimpleImputer(strategy="median"))])
cat_var_mode_imputer=Pipeline(steps=[("imputer",SimpleImputer(strategy="most_frequent"))])
cat_var_const_imputer=Pipeline(steps=[("imputer",SimpleImputer(strategy="constant",fill_value="missing"))])
```

```python
preprocessor=ColumnTransformer(transformers=[("mean_impute",num_var_mean_imputer,num_var_mean),
                              ("median_imputer",num_var_median_imputer,num_var_median),
                              ("mode_imputer",cat_var_mode_imputer,cat_var_mode),
                              ("const_imputer",cat_var_const_imputer,cat_var_const)]
                )
#syntax == ColumnTransformer(transformers=[("Name",Pipeline_name,Column_name)]
```

```python
preprocessor.fit(X_train2)
```

```python
preprocessor.transform
```

```python
X_train2_clean=preprocessor.transform(X_train2)
X_test_clean=preprocessor.transform(test2)
```

```python
#convert array to dataframe with all cleaned columns 
X_train2_clean_miss_var=pd.DataFrame(X_train2_clean,columns=num_var_mean+num_var_median+cat_var_mode+cat_var_const)
```
